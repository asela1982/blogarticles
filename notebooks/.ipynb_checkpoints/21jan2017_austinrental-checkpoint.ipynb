## 1. Develop the python script and scrape the information

```python
# import dependencies
import time
from bs4 import BeautifulSoup
from splinter import Browser
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#to get the full path to the chromedriver
!which chromedriver
```

```python
# To use the Chrome driver, 
executable_path = {'executable_path':'/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path)
```

```python
# visit the craiglist site using the browser object from splinter
url = 'https://austin.craigslist.org/search/apa'
browser.visit(url)
time.sleep(1)
```

```python
# scrape the page into soup
html = browser.html
soup = BeautifulSoup(html,'lxml')
```

```python
# sample one block that contains the elements that contains the sought info to identify the selector paths
soup.find_all('li',class_='result-row')[0]
```

```python
# once scanning the above as well as inspecting the elements in the terminal we can decide on the selector paths
# to the specific elements that contains sought info
# title : sortable-results > ul > li:nth-child(1) > p > a
# price : sortable-results > ul > li:nth-child(1) > p > span.result-meta > span.result-price
# floorarea: sortable-results > ul > li:nth-child(1) > p > span.result-meta > span.housing
# hood : sortable-results > ul > li:nth-child(1) > p > span.result-meta > span.result-hood
# sortable-results > ul > li:nth-child(112) > p > span.result-meta > span.housing
```

```python
# Grab all li element blocks
results = soup.find_all('li',class_='result-row')
```

```python
# Using splinter Click the next button
# css selector for the next button : div.paginator.buttongroup span.buttons a.button.next
browser.find_by_css("div.paginator.buttongroup span.buttons a.button.next").click()
time.sleep(5)
```

```python
# putting everything together

# empty list to populate the scrapping results
mylist = []

# Loop through ten pages
for x in range(10):
    
    html = browser.html
    soup = BeautifulSoup(html,'lxml')
    results = soup.find_all('li',class_='result-row')

    for result in results: 
        
        # an empty dictionary to hold the values obtained from each iteration
        mydict = {}  
        # Error-handling
        try:
            title = result.p.a.get_text()
            price = result.p.find('span',class_='result-meta').find('span',class_='result-price').get_text()
            area = result.p.find('span',class_='result-meta').find('span',class_='housing').get_text()
            hood = result.p.find('span',class_='result-meta').find('span',class_='result-hood').get_text()
      
            # scrapped results inserted to the empty dictionary
            mydict = {
                'title':title,
                'price':price,
                'area':area,
                'hood':hood
                    }
            
            # append the results to the mylist
            mylist.append(mydict)
            
        except AttributeError as e:
            pass

    # Using splinter Click the next button
    # css selector for the next button : div.paginator.buttongroup span.buttons a.button.next
    browser.find_by_css("div.paginator.buttongroup span.buttons a.button.next").click()
    time.sleep(10)
```

## 2. Insert the scrapped results into a pandas dataframe

```python
# convert the list of dictionaries into a dataframe
df = pd.DataFrame(mylist)
```

```python
# inspect the dataframe
display(df.head())
display(df.shape) 

# we have scrapped information on 1074 rental properties!
```

## 3. Preprocessing and feature engineering

```python
# extract the square feet value
df['area_ft2']=df['area'].str.extract('(\d+)ft', expand=False)
```

```python
# extract the number of bed rooms
df['bedrooms']=df['area'].str.extract('(\d+)br', expand=False)
```

```python
# extract the dollar price
df['price_dollar']=df['price'].str.extract('\$(\d+)', expand=False)
```

```python
# extract the neighbourhood
df['neighbourhood']=df['hood'].str.extract('\((.*?)\)', expand=False)
# convert the neighbourhood to lowercase
df['neighbourhood'] = df['neighbourhood'].str.replace(' ','').str.lower()
```

```python
# drop columns that are not required
df = df.drop(['area', 'hood','price'], axis=1)
```

```python
# inspect the dataframe
display(df.head())
display(df.shape)
```

```python
# lets rename northwestaustin to northwest
df['neighbourhood'] = df['neighbourhood'].str.replace('northwestaustin','northwest')
```

```python
# obtain the neighbourhoods having atleast 5 data records
condition = df.neighbourhood.value_counts().reset_index()
neighbourhood = condition[condition['neighbourhood']>30]['index']
neighbourhood
```

```python
# drop rows corresponding to the neighbourhood austin
df = df[df['neighbourhood'] !="austin"]
```

```python
# filter the dataframe to include only the data records corresponding to the selected neighbourhoods above
df = df[df['neighbourhood'].isin(neighbourhood)]
display(df.head())
display(df.shape)
```

```python
# get the price per bedroom
df['price_per_br'] = pd.to_numeric(df['price_dollar'])/pd.to_numeric(df['bedrooms'])
```

```python
display(df.head())
display(df.shape)
```

## 4. Visualize the findings

```python
summary = df.groupby(['neighbourhood']).agg({'price_per_br':np.mean}).reset_index()
summary = summary.sort_values('price_per_br',ascending=False)
summary
```

```python
plt.figure(figsize=(15,7))
sns.barplot(x='neighbourhood',y='price_per_br',data=summary,palette="Set2")
_ = plt.ylabel('Price per Bedroom')
_ = plt.title('Average Price Per Bedroom')
_ = plt.xlabel('Neighbourhood')
plt.savefig("../Resouces/images/21jan2017/Average Price Per Bedroom.png")
plt.savefig("../Resouces/images/21jan2017/Average Price Per Bedroom.svg")
plt.show()
```

## 5. Bonus : Generate Word clouds on the title text for each Neighbourhoods

```python
dict = {}
list = []


for row,columuns in df.iterrows():
    if columuns['neighbourhood'] in dict:
        dict[columuns['neighbourhood']] += columuns['title']
    else:
         dict[columuns['neighbourhood']] = list
```

```python
mydict = {}

for row,columuns in df.iterrows():
    if columuns['neighbourhood'] in mydict:
        # append the new number to the existing array at this slot
        mydict[columuns['neighbourhood']].append(columuns['title'])
    else:
        # create a new array in this slot
        mydict[columuns['neighbourhood']] = [columuns['title']]
```

```python
mydict
```

```python
mylistvalues = []
mylistkeys = []

for key,value in mydict.items():
    new = '-'.join(value)
    mylistkeys.append(key)
    mylistvalues.append(new)
new_dict = {k: v for k, v in zip(mylistkeys, mylistvalues)}
new_dict
```
